```python
import matplotlib.pyplot as plt
import edward as ed
import numpy as np
import tensorflow as tf
from edward.models import Bernoulli,RelaxedBernoulli, RelaxedOneHotCategorical
from edward.models.random_variables import TransformedDistribution
from scipy.stats import norm,lognorm
import math
import time
```

```python
sess=ed.get_session()
tf.global_variables_initializer().run()
samples=sess.run(RelaxedBernoulli(0.05,probs=0.4,sample_shape=100000))
#samples=sess.run(Bernoulli(0.8,sample_shape=100000))
print(np.mean(samples))
```

```python
fig = plt.figure()
plt.xlabel("$x$")
plt.hist(samples,bins=100)
plt.show()
```
