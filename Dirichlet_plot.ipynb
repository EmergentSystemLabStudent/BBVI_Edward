```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
import scipy.stats as ss
```

```python
alphas = np.array([3.0, 13.0, 4.0])
dc = ss.dirichlet(alphas)     
X = []
Y = []
for x in np.arange(0.01,0.99,0.01):
    for y in  np.arange(0.01,0.99-x,0.01):
        X.append(x)
        Y.append(y)
Z = []
for x,y in zip(X,Y):
        Z.append(dc.pdf((x,y)))
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlim(0.0, 10.0)
surf = ax.plot_trisurf(X, Y, Z, cmap=cm.rainbow, linewidth=0, antialiased=False)
x = [0.33,0.33]
y = [0.33,0.33]
z = [0.0,10.0]
ax.plot(x, y, z)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
